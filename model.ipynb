```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import metrics
```

```python
data = pd.read_csv("datasets.csv")
```

```python
data.head()
```

```python
def features_to_points(col_data):
    for i,text in enumerate(col_data):
        splitted_text = text.split(';')
        if "None" in splitted_text:
            col_data[i] = 0
        else:
            col_data[i] = len(splitted_text)
features_to_points(data['friends_issues'])
features_to_points(data['family_issues'])
features_to_points(data['financial_issues'])
features_to_points(data['health_issues'])
```

```python
data.head()
```

```python
def one_or_zero(col_data):
    for i,text in enumerate(col_data):
        if (text == 'Male') or (text=='Yes'):
            col_data[i] = 1
        else:
            col_data[i] = 0
one_or_zero(data['Gender'])
one_or_zero(data['overload'])
one_or_zero(data['unpleasant'])
one_or_zero(data['academic'])
one_or_zero(data['career'])
one_or_zero(data['criticism'])
one_or_zero(data['conflicts'])
```

```python
data.head()
```

```python
data['stress_level']=data['stress_level'].replace({'Acute Stress':0,
'Episodic acute stress':1,
'Chronic stress': 2})
```

```python
y = data['stress_level']
X = data.drop("stress_level",axis=1)
```

```python
data['stress_level'].unique()
```

```python
X
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
```

```python
clf = RandomForestClassifier()
clf.fit(X_train,y_train)
```

```python
y_pred = clf.predict(X_test)
```

```python
print("Accuracy: ",metrics.accuracy_score(y_test,y_pred))
```

```python
import pickle 
pickle.dump(clf,open('model.pkl','wb'))
```
